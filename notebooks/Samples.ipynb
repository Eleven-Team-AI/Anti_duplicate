```python
companies = pd.read_csv('/Users/voronik/DeepLearning/Anti_duplicate/cleaned.csv')
```

```python
companies
```

```python
non_duplicate = companies[companies['is_duplicate']==0]
```

```python
duplicate = companies[companies['is_duplicate']==1]
```

```python
duplicate
```

```python
training_data_nonduplicate = non_duplicate.sample(frac=0.8, random_state=25)
testing_data_nonduplicate = non_duplicate.drop(training_data_nonduplicate.index)

training_data_duplicate = duplicate.sample(frac=0.8, random_state=25)
testing_data_duplicate = duplicate.drop(training_data_duplicate.index)
```

```python
train = pd.concat([training_data_nonduplicate, training_data_duplicate])
test = pd.concat([testing_data_nonduplicate, testing_data_duplicate])
```

```python
train.to_csv('train.csv')
test.to_csv('test')
```
